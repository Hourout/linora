# Image Augmentation

### In the linora module, there are two image enhancement methods, 
### one is to use la.image.xx, such as la.image.RandomContrast(), 
### the other is to use la.image.ImageAug(), through chain calculations can be faster calculations and experiments.

```python
image = la.image.read_image('yichan_ok.jpg')
image = la.image.RandomFlipLeftRight(image, random=False)
Image.fromarray(image)
```

```python
la.image.ImageAug().read_image('yichan_ok.jpg').RandomFlipLeftRight(random=False).show()
```

```python
la.image.ImageAug().read_image('yichan_ok.jpg').RandomFlipLeftRight(random=False).run()
```

```python
la.image.ImageAug(image).RandomFlipLeftRight(random=False).show()
```
